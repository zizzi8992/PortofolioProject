```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
netflix_df = pd.read_csv('netflix_raw.csv')
netflix_df
```

```python
netflix_df = netflix_df.drop(['description'], axis=1)
```

```python
netflix_df.head()
```

```python
netflix_df.tail()
```

```python
netflix_df.shape
```

```python
#alter the column listed_in to category
netflix_df.rename(columns={'listed_in' : 'category'}, inplace=True)
netflix_df.head()
```

```python
netflix_df.describe()
```

Summary dari data netflix_df hanya berupa release_year karena hanya kolom ini yang dtypes nya int, sedangkan yang lain merupakan categorical dtypes

```python
netflix_df.info()
```

```python
#Cheking null data
netflix_df.isnull().sum()
```

```python
netflix_df = netflix_df.dropna()
netflix_df
```

```python
netflix_df.isnull().sum()
```

```python
#checking is there any duplicated data (there is none)
netflix_df.duplicated().value_counts()
```

```python
#checking unique value of type
netflix_df['type'].nunique()
netflix_df['type'].unique()
```

```python
#change the type TV Show into Series
netflix_df['type'] = netflix_df['type'].replace('TV Show', 'Series')
netflix_df['type']
```

```python
#value count for each of "type", "country", "rating",and "director" columns
cols = ["type", "country", "rating","director"]
for col in cols:
  print(netflix_df[col].value_counts(), "\n")
```

```python
#The ammount of Movies and Series
sns.countplot(netflix_df, y='type')
```

```python
#Top 15 directors
top_15_dir = netflix_df['director'].value_counts().head(15)
top_15_dir
```

```python
plt.figure(figsize = (15, 7))
sns.barplot(x = top_15_dir.values, y = top_15_dir.index)
plt.title("Top 15 directors")
plt.show()
```

```python
#Top 15 countries
top_15_countries = netflix_df['country'].value_counts().head(15)
top_15_countries
```

```python
plt.figure(figsize = (15, 7))
sns.barplot(x = top_15_countries.values, y = top_15_countries.index)
plt.title("Top 15 countries")
plt.show()
```

```python
#The Rating Counts
rating = netflix_df['rating'].value_counts()
rating
```

```python
changename = {
'TV-MA':'Mature',
'TV-14':'Kids over 14',
'TV-PG':'Parental Guidance Suggested',
'R':'Restricted',
'PG-13':'Parents Strongly Cautioned',
'TV-Y7':'Kids over 7',
'TV-Y':'Kids of all ages',
'PG':'Parental Guidance Suggested',
'TV-G':'All ages',
'NR':'Not Rated (uncut version)',
'G':'General Audiences',
'TV-Y7-FV':'Contain Fantasy Violence',
'NC-17':'Adults Only',
'UR':'Unrated'
}

netflix_df['rating'] = netflix_df['rating'].replace(changename)
netflix_df['rating']

rating = netflix_df['rating'].value_counts().head(15)
rating
```

```python
plt.figure(figsize = (15, 10))
sns.barplot(x = rating.values, y = rating.index)
plt.title("The Rating Counts")
plt.show()
```

```python
#Top 15 categories
top_15_genre = netflix_df['category'].value_counts().head(15)
top_15_genre
```

```python
plt.figure(figsize = (12, 7))
sns.barplot(x = top_15_genre.values, y = top_15_genre.index)
plt.title("Top 15 categories")
plt.show()
```

```python
#Most Popular Cast
netflix_df['cast']=netflix_df['cast'].str.split(", ") # means column cast the content will split using - as delimeter
netflix_df=netflix_df.explode("cast",ignore_index=True)
netflix_df
```

```python
netflix_df['cast'].isnull().sum()
```

```python
#Top 15 cast
top_15_cast = netflix_df['cast'].value_counts().head(15)
top_15_cast
```

```python
plt.figure(figsize = (12, 7))
sns.barplot(x = top_15_cast.values, y = top_15_cast.index)
plt.title("Top 15 cast")
plt.show()
```
